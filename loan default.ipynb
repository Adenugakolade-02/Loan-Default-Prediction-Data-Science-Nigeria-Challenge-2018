#### For some feature engineering

```python
prev['closeddate'] = pd.to_datetime(prev['closeddate'])
prev['firstduedate'] = pd.to_datetime(prev['firstduedate'])
prev['firstrepaiddate'] = pd.to_datetime(prev['firstrepaiddate'])
```

```python
prev['closeddate'] = [datetime.timestamp(i) for i in prev['closeddate']]
prev['firstduedate'] = [datetime.timestamp(i) for i in prev['firstduedate']]
prev['firstrepaiddate'] = [datetime.timestamp(i) for i in prev['firstrepaiddate']]
```

```python
prev = pd.DataFrame(prev, columns=['customerid','closeddate','firstduedate','firstrepaiddate'])
```

```python
prev.head()
```

```python
pf_data.head()
```

```python
pf_data.loc[pf_data['referredby'].notnull(), 'referredby'] = 1
pf_data.loc[pf_data['referredby'].isnull(), 'referredby'] = 0
```

```python
train = pd.merge(pf_data,prev,on='customerid',how='inner')
```

```python
train.head()
```

```python
train.loc[train['good_bad_flag']=='Good','good_bad_flag'] = 1
train.loc[train['good_bad_flag']=='Bad','good_bad_flag'] = 0
```

```python
train['approveddate'] = pd.to_datetime(train['approveddate'])
train['creationdate'] = pd.to_datetime(train['creationdate'])
train['customerid'] = train['customerid'].astype('category')
```

```python
train['approveddate'] = [datetime.timestamp(i) for i in train['approveddate']]
train['creationdate'] = [datetime.timestamp(i) for i in train['creationdate']]
```

```python
train = train.groupby(train['customerid']).mean().reset_index()
```

```python
target = train['good_bad_flag']
train = train.drop('customerid',axis=1)
del(train['good_bad_flag'])
```

```python
#Scaling the data, so as to increase the computational speed
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
train.head()
```

```python
on = ['systemloanid','loannumber','approveddate','creationdate','loanamount','totaldue','termdays','referredby','closeddate','firstduedate','firstrepaiddate']
```

```python
train[on] = scale.fit_transform(train[on])
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
X_train,X_test,y_train,y_test = train_test_split(train,target,test_size=0.25,random_state=42)
```

```python
forest2 = RandomForestClassifier(n_estimators=40, max_features=3,max_leaf_nodes=5,random_state=0)
forest2.fit(X_train,y_train)
```

```python
print('the RandomForestClassifier 1-accuracy score of train set is: {}'.format(1-(accuracy_score(y_train,forest2.predict(X_train)))))
print('the RandomForestClassifier 1-accuracy score of test set is: {}'.format(1-(accuracy_score(y_test,forest2.predict(X_test)))))
```
